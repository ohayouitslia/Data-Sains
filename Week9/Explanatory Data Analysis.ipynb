```
# Disney Movies Data Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load the Disney movies data
df = pd.read_csv('disney_movies_total_gross.csv')
```

```
# Process the data
df.columns = df.columns.str.strip()
df['release_date'] = pd.to_datetime(df['release_date'])
df['Year'] = df['release_date'].dt.year
```

```
# Sort by Total Gross and take top 10
top_10 = df.sort_values(by='total_gross', ascending=False).head(10)
```

```
# Visualization 1: Top 10 Disney Movies by Total Gross
plt.figure(figsize=(12, 6))
sns.barplot(x='total_gross', y='movie_title', data=top_10, palette='Blues_d')
plt.title('Top 10 Disney Movies by Total Gross')
plt.xlabel('total_gross')
plt.ylabel('movie_title')
plt.tight_layout()
plt.show()
```

```
# Visualization 2: Total Gross Over Time
plt.figure(figsize=(14, 6))
sns.lineplot(data=df.groupby('release_date')['total_gross'].sum().reset_index(), x='release_date', y='total_gross')
plt.title('Total Gross of Disney Movies Over Time')
plt.xlabel('release_date')
plt.ylabel('total_gross')
plt.tight_layout()
plt.show()
```

```
# Visualization 3: Show Total Gross by MPAA Rating
if 'total_gross' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='mpaa_rating', y='total_gross', hue='Year', data=df, palette='viridis', alpha=0.6)
    plt.title('Number of Disney Movies by MPAA Rating')
    plt.xlabel('mpaa_rating')
    plt.ylabel('total_gross')
    plt.tight_layout()
    plt.show()
```

```
# Visualization 5: Bar Chart - Average Total Gross by Genre (Top 10 Genres)
if 'genre' in df.columns:
    top_genres = df.groupby('genre')['total_gross'].mean().sort_values(ascending=False).head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(x=top_genres.index, y=top_genres.values, palette='Set2')
    plt.title('Average Total Gross by Top 10 Genres')
    plt.xlabel('genre')
    plt.ylabel('total_gross')
    plt.tight_layout()
    plt.show()
```

```
# Visualization 6: Top 10 Disney Movies by Total Gross
top_10 = df.sort_values(by='total_gross', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='total_gross', y='movie_title', data=top_10, palette='Oranges_d')
plt.title('Top 10 Disney Movies by Total Gross')
plt.xlabel('total_gross')
plt.ylabel('movie_title')
plt.tight_layout()
plt.show()
```
